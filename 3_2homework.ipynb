```python
import pymongo 
import charts
```

```python
client = pymongo.MongoClient('localhost',27017)
ceshi = client['ceshi']
sample_info = ceshi['sample_info']
```

```python
for i in sample_info.find().limit(300):
    print(i)
```

```python
cates_list = []
for i in sample_info.find():
    cates_list.append(i['cates'][2])
cates_index = list(set(cates_list))   #set去重
print(cates_index,len(cates_index))
```

```python
post_times = []
for index in cates_index:
    post_times.append(cates_list.count(index))
print(post_times)
```

```python
def data_gen(types):
    for cate,times in zip(cates_index,post_times):
        data = {
            'name':cate,
            'data':[times],
            'type':types
        }
        yield data
```

```python
for i in data_gen('column'):
    print(i)
```

```python
series = [data for data in data_gen('column')]
charts.plot(series,show='inline',options=dict(title=dict(text='发帖量')))
```
